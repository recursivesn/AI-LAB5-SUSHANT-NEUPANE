**k-Means Clustering for 2 Features with k=2 Clusters**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Dataset
data = np.array([
    [1.0, 1.5],
    [1.5, 1.8],
    [1.2, 1.3],
    [1.4, 1.7],
    [5.0, 5.5],
    [5.5, 5.8],
    [5.2, 5.3],
    [5.4, 5.7],
    [2.0, 2.1],
    [1.8, 1.9],
    [5.1, 5.4],
    [5.3, 5.6]
])

# Apply KMeans clustering with k=2
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(data)

# Get cluster assignments and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Compute Within-Cluster Sum of Squares (WCSS)
wcss = kmeans.inertia_

# Print results
print("Cluster Assignments:", labels)
print("Centroids:\n", centroids)
print("WCSS (Within-Cluster Sum of Squares):", wcss)

# Plotting the clusters and centroids
plt.figure(figsize=(8, 6))
for i in range(2):
    cluster_points = data[labels == i]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {i+1}')

# Plot centroids
plt.scatter(centroids[:, 0], centroids[:, 1], color='black', marker='X', s=200, label='Centroids')

# Plot details
plt.title('k-Means Clustering (k=2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

**k-Means Clustering for Arbitrary Features with k Clusters**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Sample dataset with 3 features (you can add more rows or columns)
data = np.array([
    [1.0, 1.5, 0.5],
    [1.5, 1.8, 0.6],
    [1.2, 1.3, 0.7],
    [1.4, 1.7, 0.8],
    [5.0, 5.5, 5.1],
    [5.5, 5.8, 5.3],
    [5.2, 5.3, 5.2],
    [5.4, 5.7, 5.4],
    [2.0, 2.1, 1.9],
    [1.8, 1.9, 2.0],
    [5.1, 5.4, 5.5],
    [5.3, 5.6, 5.6]
])

# Set number of clusters
k = 3

# Run k-means
kmeans = KMeans(n_clusters=k, random_state=0)
kmeans.fit(data)

# Results
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
wcss = kmeans.inertia_

# Print results
print("Cluster labels:", labels)
print("Centroids:\n", centroids)
print("WCSS:", wcss)

# Simple 2D Plot (only using first 2 features)
plt.figure(figsize=(8, 6))
for i in range(k):
    cluster = data[labels == i]
    plt.scatter(cluster[:, 0], cluster[:, 1], label=f'Cluster {i+1}')

plt.scatter(centroids[:, 0], centroids[:, 1], color='black', marker='X', s=200, label='Centroids')
plt.title("k-Means Clustering (k=3)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```
